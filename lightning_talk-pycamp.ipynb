### Drumhead example 

Bessel functions are a family of solutions to Bessel's differential equation:
$$x^2 \frac{d^2}{dx^2} + x\frac{dy}{dx} + (x^2 - \alpha^2)y = 0$$
These show up when solving for vibrational modes.

```python
from scipy.special import jn, jn_zeros

def drumhead_height(n, k, distance, angle, t):
   nth_zero = jn_zeros(n, k)
   return np.cos(t) * np.cos(n*angle) * jn(n, distance*nth_zero)

theta = sp.r_[0:2*np.pi:50j]
radius = sp.r_[0:1:50j]
x = np.array([r*np.cos(theta) for r in radius])
y = np.array([r*np.sin(theta) for r in radius])
z = np.array([drumhead_height(1, 1, r, theta, 0.5) for r in radius])
```

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
fig = pylab.figure()
ax = Axes3D(fig)
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.jet)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
pylab.show()
```
